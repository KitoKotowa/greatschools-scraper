```python

# create city state list
cities = pd.read_csv('uscities.csv')

# just get the second and third colun
cities = cities[['city','state_name','state_id']]

# create new dictionary with reversed key, value pairs
# making sure state/city combo conform to url format of "-" for " "
cities['state_name'] = cities['state_name'].str.strip()
cities['state_name'] = cities['state_name'].str.replace(" ", "-")
cities['city'] = cities['city'].str.replace(" ", "-")
# remove extraneous header rows
cities['city'] = (cities['state_name'] + '/'+ cities['city']).str.lower()
print(cities.head())

# persist by creating new csv
cities.to_csv('cities.csv')
```
